```python
import time

def extract_single_page(image_path):
    start = time.time()

    result = ocr.ocr(image_path)

    items = []

    for item in result[0]:
        box = item[0]
        text = item[1][0]
        conf = item[1][1]

        x = box[0][0]
        y = box[0][1]

        items.append((x, y, text, conf))

    items = sorted(items, key=lambda x: (x[1], x[0]))

    lines = []
    conf_weighted = []
    total_chars = 0

    for _, _, text, conf in items:
        lines.append(text)
        conf_weighted.append(conf * len(text))
        total_chars += len(text)

    raw_text = "\n".join(lines)
    formatted_text = "\n\n".join(lines)

    confidence = sum(conf_weighted) / total_chars if total_chars else 0

    elapsed = time.time() - start

    return {
        "status": "success",
        "extracted_text": raw_text,
        "formatted_text": formatted_text,
        "confidence_score": float(confidence),
        "metadata": {
            "pages": 1,
            "language_detected": "en",
            "processing_time_sec": round(elapsed, 2)
        }
    }
```

```python
res = extract_single_page(r"C:\Users\Asus\OneDrive\Pictures\Screenshots\sample written english.jpg")
print(res["formatted_text"])
```

## PDF Extraction

```python
from pdf2image import convert_from_path
import numpy as np
import time

POPPLER_PATH = r"C:\Users\Asus\Downloads\ocr-project\poppler\poppler-25.12.0\Library\bin"

def extract_pdf(pdf_path):
    start = time.time()

    pages = convert_from_path(pdf_path, dpi=300, poppler_path=POPPLER_PATH)

    all_lines = []
    conf_weighted = []
    total_chars = 0

    for page in pages:
        # IMPORTANT FIX → PIL → numpy
        result = ocr.ocr(np.array(page))

        items = []

        for item in result[0]:
            box = item[0]
            text = item[1][0]
            conf = item[1][1]

            x = box[0][0]
            y = box[0][1]

            items.append((x, y, text, conf))

        items = sorted(items, key=lambda x: (x[1], x[0]))

        for _, _, text, conf in items:
            all_lines.append(text)
            conf_weighted.append(conf * len(text))
            total_chars += len(text)

    raw_text = "\n".join(all_lines)
    formatted_text = "\n\n".join(all_lines)

    confidence = sum(conf_weighted) / total_chars if total_chars else 0

    elapsed = time.time() - start

    return {
        "status": "success",
        "extracted_text": raw_text,
        "formatted_text": formatted_text,
        "confidence_score": float(confidence),
        "metadata": {
            "pages": len(pages),
            "language_detected": "en",
            "processing_time_sec": round(elapsed, 2)
        }
    }
  
```

```python
res = extract_pdf(r"C:\Users\Asus\Downloads\ocr-project\sample datas\sample pdf.pdf")

import json
print(json.dumps(res, indent=2))
```

```python
import re

def parse_invoice_fields(text: str):

    def find(patterns):
        for p in patterns:
            m = re.search(p, text, re.IGNORECASE)
            if m:
                return m.group(1).strip()
        return None

    return {
        "invoice_number": find([
            r"invoice\s*(no|number)[\s:]*([A-Z0-9\-]+)",
            r"inv[\s\-]*([A-Z0-9\-]+)"
        ]),

        "date": find([
            r"date[\s:]*([A-Za-z0-9 ,\-]+)"
        ]),

        "total_amount": find([
            r"total\s*(amount)?[\s:₹]*([0-9,\.]+)",
            r"grand\s*total[\s:₹]*([0-9,\.]+)"
        ]),

        "gst": find([
            r"gst.*?([0-9,\.]+)",
            r"tax.*?([0-9,\.]+)"
        ]),

        "account_number": find([
            r"account\s*(number)?[\s:]*([0-9]{6,})"
        ]),

        "ifsc": find([
            r"ifsc[\s:]*([A-Z0-9]+)"
        ])
    }
```

```python
res = extract_single_page(r"C:\Users\Asus\OneDrive\Pictures\Screenshots\Screenshot 2025-11-10 210059.png")

fields = parse_invoice_fields(res["formatted_text"])
print(fields)
```
